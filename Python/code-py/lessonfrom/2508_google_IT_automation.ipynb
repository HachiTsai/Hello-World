````python
# https://www.coursera.org/learn/python-crash-course/supplement/A30Rl/tuples

def convert_seconds(seconds):
    hours = seconds // 3600
    minutes = (seconds - hours * 3600) // 60
    remaining_seconds = seconds - hours * 3600 - minutes * 60
    return hours, minutes, remaining_seconds

result = convert_seconds(5000)
type(result) # <class 'tuple'>

print(result)  # (1, 23, 20)

hours, minutes, seconds = result  # 解包元組
print(hours, minutes, seconds)  # 1 23 20

hours, minutes, seconds = convert_seconds(1000)  # 解包元組
print(hours, minutes, seconds)  # 0 16 40
````

````python
import shutil
import psutil

def check_disk_usage(disk):
    du = shutil.disk_usage(disk)
    free = du.free / du.total * 100
    return free  > 20 # Percentage of free disk space

def check_cpu_usage():
    usage = psutil.cpu_percent(1)
    return usage < 75

if not check_disk_usage("/") or not check_cpu_usage():
    print("ERROR!")
else:
    print("Everything is OK!")

````

以下是針對[IT問題排查與除錯]的系統化方法指南，結合業界實務經驗與技術深度解析，幫助你建立全面的問題解決框架：  

---  

### **1. 問題分類與對應策略**  
#### **四大核心問題類型**  
| **問題類型**               | **典型場景**                          | **工具/方法**                          |  
|----------------------------|---------------------------------------|----------------------------------------|  
| **程序崩潰 (Crash)**        | Segmentation Fault、Python Traceback | `gdb` (C)、`pdb` (Python)、日誌分析    |  
| **效能瓶頸 (Performance)**  | CPU 100%、記憶體洩漏                  | `top`、`perf`、`cProfile` (Python)     |  
| **邏輯錯誤 (Logic Error)**  | 輸出結果不符合預期                     | 單元測試、斷點除錯、`print()` 調試    |  
| **系統整合問題**            | 權限拒絕、網路不通                     | `strace`、`tcpdump`、系統日誌分析      |  

---  

### **2. 通用除錯流程 (五步法)**  
```mermaid  
graph TD  
    A[現象觀察] --> B[假設建立]  
    B --> C[實驗驗證]  
    C --> D[根本原因分析]  
    D --> E[解決方案實施]  
```  

#### **步驟詳解**  
1. **現象觀察**  
   - 收集錯誤訊息（完整 Traceback）  
   - 重現步驟（最小化重現環境）  
   - 使用 `journalctl -xe` (Linux) 或 Event Viewer (Windows) 查系統日誌  

2. **假設建立**  
   - 對照已知模式（例如：`Errno 13` 通常為權限問題）  
   - 使用 **橡皮鴨除錯法** 口頭描述問題  

3. **實驗驗證**  
   - 隔離測試：用 `docker run --rm` 建立乾淨環境測試  
   - 對照實驗：修改單一變數觀察差異  

4. **根本原因**  
   - 時間軸分析：比對程式行為與系統事件時間戳  
   - 依賴檢查：`ldd` (Linux) 或 `Dependency Walker` (Windows)  

5. **解決方案**  
   - 短期修復：重啟服務、回滾版本  
   - 長期方案：修補程式碼、增加監控  

---

### **4. 實戰案例解析**  
#### **案例一：服務無響應**  
- **現象**：Nginx 返回 502 Bad Gateway  
- **排查步驟**：  
  1. 檢查上游服務日誌：`journalctl -u gunicorn`  
  2. 確認連線性：`telnet localhost 8000`  
  3. 分析請求阻塞：`strace -p <PID>`  
- **根因**：資料庫連線池耗盡  
- **解決**：調整 PostgreSQL `max_connections` 參數  

#### **案例二：Python 記憶體暴漲**  
- **診斷工具**：

  ```  
- **常見原因**：  
  - 未關閉檔案描述符  
  - 全域變數累積資料

### 1. **問題解決的三階段流程**  
#### (1) 資訊收集（Gathering Information）  
- **目標**：全面理解問題現象與背景。  
- **重點方法**：  
  - **問題描述**：記錄[發生時機、頻率、觸發條件]（如用戶操作時電腦無預警關機）。  
  - **資源利用**：  
    - 系統日誌（如`/var/log`中的錯誤訊息）。  
    - 內部文件、網路論壇（如Stack Overflow）。  
  - **重現案例**：建立可重複觸發問題的步驟（關鍵於間歇性問題）。  

#### (2) 根因分析（Root Cause Analysis）  
- **目標**：定位問題源頭，而非僅處理表面症狀。  
- **重點工具與技巧**：  
  - **日誌分析**：例如發現[溫度閾值超標]錯誤，指向過熱問題。  
  - **硬體檢查**：實際拆機發現CPU風扇被灰塵阻塞。  
  - **二元搜尋法**：逐步排除可能性（如先區分軟體/硬體問題）。  

#### (3) 修復與預防（Remediation）  
- **目標**：分階段實施解決方案。  
- **層次策略**：  
  - **短期修復**：立即恢復服務（如清潔風扇）。  
  - **長期預防**：  
    - 部署監控系統（如溫度異常警報）。  
    - 環境改善（如減少機房灰塵）。  

---

- **實際挑戰**：

#### **系統呼叫分析技巧**  
```bash

#### **解讀關鍵日誌行**  
| **部分**               | **意義**                                                                 |  
|------------------------|--------------------------------------------------------------------------|  
| `openat`               | 系統呼叫類型（嘗試打開文件/目錄）                                        |  
| `O_DIRECTORY`          | 標誌位表示嘗試打開的是目錄                                               |  
| `-1 ENOENT`            | 返回值為-1，錯誤碼 `ENOENT` 表示目標不存在                               |

## [問題追蹤] Purple Box 啟動失敗 (版本 2.1.0)  

### 現象  
- 升級至 v2.1.0 後應用程式無錯誤退出  
- 影響所有 Linux 用戶  

### 根因分析  
程式啟動時未檢查 `~/.config/purple-box` 目錄存在性，導致 `openat()` 系統呼叫失敗  

### 解決方案  
1. 臨時措施：手動創建目錄  
   ```bash  
   mkdir -p ~/.config/purple-box  
   ```  
2. 永久修復：已提交熱修補 (PR #123) 新增目錄自動創建邏輯  

### 預防措施  
- 在 CI 流程中加入[裸機環境測試]階段  
- 更新開發者檢查清單（必要目錄驗證）  
```

### **總結：除錯心法**  
1. **從靜默失敗中找線索**：無錯誤訊息時優先檢查系統呼叫 (`strace`) 和日誌 (`journalctl`)  
2. **最小化重現**：在新用戶帳號或 Docker 容器中測試可快速排除環境干擾  
3. **橫向思考**：同類問題可能隱藏在其他路徑檢查（如 `/tmp/.purple-box`）  

> 💡 **專家提示**：當 `strace` 輸出過多時，可用 `-e trace=file` 只追蹤文件相關系統呼叫，或結合 `-f` 參數追蹤子進程！
